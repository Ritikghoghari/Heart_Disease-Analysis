# 11.Find Count of Male & Female in this Dataset


```python
df.columns
```

```python
df['sex'].value_counts()

```

```python
sns.countplot(x = df['sex'] , palette='magma')
plt.xticks([0,1],['Female','Male'])
plt.show()
```

```python
gender = df['sex'].value_counts()
explodes = [0.1,0]


plt.figure(figsize=(5, 10))
plt.pie(gender,explode=explodes ,labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'],shadow=True)
plt.title("Gender Distribution Men Vs Female")
plt.show()
```
